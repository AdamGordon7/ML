```python
def f(x):
    return (x**2)*(np.sin(2*np.pi*x))+0.7

def gen_points(size):
    return np.random.uniform(low=0,high=1,size=(size,2))

def classify_points(dataset):
    class_0=[]
    class_1=[]
    for i in dataset:
        if(f(i[0])>i[1]):
            class_0.append(i)
        else:
            class_1.append(i)
    
    return(class_0,class_1)

def plot_points(x,y):
    x_0=[]
    y_0=[]

    x_1=[]
    y_1=[]

    for i in class_0:
        x_0.append(i[0])
        y_0.append(i[1])

    for i in class_1:
        x_1.append(i[0])
        y_1.append(i[1])

    plt.plot(x_0,y_0,"o")
    plt.plot(x_1,y_1,"x")
    plt.show()


dataset=gen_points(100)
class_0,class_1=classify_points(dataset)
class_0=np.array(class_0)
class_1=np.array(class_1)
print(class_0.shape)
print(class_1.shape)
plot_points(class_0,class_1)





    
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))

def init_weights(input_layer,hidden_layer,output_layer):
    theta_1=np.random.uniform(0,1,(hidden_layer,input_layer))
    theta_2=np.random.uniform(0,1,(output_layer,hidden_layer))

    return [theta_1,theta_2]


def forward_prop(thetas,data):

    class_0=data[0]
    class_1=data[1]

    class_0_activations=[]
    class_1_activations=[]
  
    for i in class_0:
        hidden_layer_z=thetas[0].dot(i)
        hidden_layer_A=sigmoid(hidden_layer_z)
     

        output_layer_z=thetas[1].dot(hidden_layer_A)
        output_layer_A=sigmoid(output_layer_z)
        class_0_activations.append(output_layer_A)

    for i in class_1:
        hidden_layer_z=thetas[0].dot(i)
        hidden_layer_A=sigmoid(hidden_layer_z)
     

        output_layer_z=thetas[1].dot(hidden_layer_A)
        output_layer_A=sigmoid(output_layer_z)
        class_1_activations.append(output_layer_A)

    return np.array(class_0_activations),np.array(class_1_activations)

class0,class1=forward_prop(init_weights(2,3,1),[class_0,class_1])
print("Class 0:")
print(class0.shape)

print("Class 1:")
print(class1.shape)


def error(activations,class_0,class_1):
    class_0_act=activations[0]
    class_1_act=activations[1]

    error0=class_0_act-class_0
    error1=class_1_act-class_1

    return error0,error1

error([class0,class1],class_0,class_1)


```
